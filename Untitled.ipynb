```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
sns.set()

lt = pd.read_csv("loan_timing.csv")
lt.describe()
```

The first step of a good data analysis is to improve our understanding of our data and where it comes from.

From the above, we observe the following:

1. Our oldest loan is 730 days. Since no loan is three years old, we don't have an estimate for the ratio of loans which get paid off. 

2. `days since origination` is approximately uniformally distributed, i.e. loans have been given at approximately a constant rate.



```python
chargeoffs = lt.dropna() 

sns.distplot(chargeoffs['days from origination to chargeoff'])
plt.show()
sns.distplot(chargeoffs['days from origination to chargeoff'],
             hist_kws=dict(cumulative=True))
plt.show()

_, counts = np.unique(chargeoffs, return_counts=True)
print(counts)
```

The majority of our chargeoffs occur earlier during the loan process, but this data is misleading since we have more data about early-stage loans than we do about later-stage ones.

A good way to account for this is to compare the chargeoffs with the number of loans that have are older than the `days from origination to chargeoff`. This is in essense the calculation of the [Survival](http://mathworld.wolfram.com/SurvivalFunction.html) and [Hazard](http://mathworld.wolfram.com/HazardFunction.html) functions.

In order calculate the survival ratio, we need to count the number of loans that have survived at each `day since origination` and the number of loans that have charged off at each point.


```python
survivors = lt.loc[lt['days from origination to chargeoff'].isna(),:].iloc[:, 0]
print(survivors.head())

ct = Counter(survivors)
prev = 0
for i in range(max(survivors), -1, -1):
    ct.setdefault(i, 0)
    ct[i] += prev
    prev = ct[i]
    
cum_surv = pd.Series(ct)
print(cum_surv.iloc[:5], '\n', cum_surv.iloc[-5:])
cum_surv.plot()
```

We calculate cumulative chargeoffs in a similar manner:

```python
#asdf = chargeoffs.loc[: , 'days from origination to chargeoff'].cumsum()[::-1]
#asdf
chargeoffs['days from origination to chargeoff'].astype(int)
ct = Counter(chargeoffs['days from origination to chargeoff'])

prev = 0
for i in np.arange(max(ct), -1, -1):
    ct.setdefault(i, 0)
    ct[i] += prev
    prev = ct[i]

cum_chrg = pd.Series(ct)
print(cum_chrg.iloc[:5], '\n', cum_chrg.iloc[-5:])
cum_chrg.plot()
```

```python
hazard = cum_chrg / (cum_chrg + cum_surv) 

hazard = pd.DataFrame({'days since origination': hazard.index, 'hazard': hazard.values})
sns.lmplot('days since origination', 'hazard', data=hazard, size=6)

sns.residplot('days since origination', 'hazard', data=hazard)
```

Our residual suggests that the hazard function is approximately quadratic.
