### Calculating _Survival_ and _Hazard_

A good way to account for this is to compare the chargeoffs with the number of loans that have are older than the `days from origination to chargeoff`. In essense, this will entail the calculation of the [Survival](http://mathworld.wolfram.com/SurvivalFunction.html) and [Hazard](http://mathworld.wolfram.com/HazardFunction.html) functions. Specifically, we will be calculating:

- The number of loans that have survived past each `day since origination` 
- The number of chargeoffs that have occured at each day.

We will represent both as `pandas.Series` objects indexed by day.


```python
survivors = lt.loc[lt['days from origination to chargeoff'].isna(),:].iloc[:, 0]
print(survivors.head())

ct = Counter(survivors)
prev = 0
for i in range(max(survivors), -1, -1):
    ct.setdefault(i, 0)
    ct[i] += prev
    prev = ct[i]
    
cum_surv = pd.Series(ct)
print(cum_surv.iloc[:5], '\n', cum_surv.iloc[-5:])
# cum_surv.plot()
```

We calculate cumulative chargeoffs in a similar manner:

```python
#asdf = chargeoffs.loc[: , 'days from origination to chargeoff'].cumsum()[::-1]
#asdf
chargeoffs['days from origination to chargeoff'].astype(int)
ct = Counter(chargeoffs['days from origination to chargeoff'])

cum_chrg = pd.Series((ct.get(i, 0) for i in cum_surv.index.values), cum_surv.index)

print(cum_chrg.iloc[:10], '\n', cum_chrg.iloc[-10:])
#cum_chrg.plot()

sns.jointplot(cum_chrg.index.values, cum_chrg.values)
```

```python
hazard = cum_chrg / (cum_chrg + cum_surv)
print(hazard.head(8), hazard.tail(8))

hazard = pd.DataFrame({'days since origination': hazard.index, 'hazard': hazard.values})

sns.jointplot('days since origination', 'hazard', data=hazard, kind='reg',
              xlim=(-14, 744), ylim=(-.0005, .003), size=8)
sns.jointplot('days since origination', 'hazard', data=hazard, kind='resid',  
              ylim=(-.0005, .003), size=8)
```

During the first ~400 days, `hazard` seems to be approximately normally-distributed accross the residual, but begins to exhibit finger-like behavior for older loans. This is caused by the fact that we have fewer loans in the older range - each of the fingers represents the number of chargebacks on the given day (zero being of course the flat one, followed by one, two, etc.). This also explains the outliers in our graph - there are fewer loans, so a single chargebacks will have a larger numeric impact on our hazard.

There are several different approaches at this point:

1) We use the 99th-percentile cutoff that we calculated earlier, and ignore all loans older than 645 days. This doesn't really work though since we'd be throwing out a lot of our rows (`days since origination` is approximately uniformally distributed), and we would induce an artificial scarcity of data near the 645-day cutoff. Since we'd be ignoring older loans, we would also lose the data form their survival history - which causes us to lose more data than we had intended. This approach eventually becomes.

2) We use the data as-is. This approach ignores the noise generated in our data, and will later manifest itself when we are measuring test and validation set errors. In the long term, this wouldn't be a bad idea since we would eventually collect enough loan data that the skewing effect becomes negligible. It's not very practical in the context of this problem though because we want to predict the chargeoffs in _this_ set of loans.

3) My proposed approach is that we use some rudimentary form of graph smoothing by binning our days into groups of say 7 days (i.e. a week). This would reduce the data sparsity / outlier effect that we see in the older loans. In this case, doing a 7-day bin is reasonable since hazard is so linear (i.e. doesn't predictably fluctuate day-to-day). Our final regression would predict information on a weekly scale, but this is also fine since we could just interpolate the weekdays if we need to, and our data already shows that this shouldn't be an issue.



```python
n_wks = 728/7
weekly_hazard = pd.DataFrame({'weeks since origination': np.arange(n_wks), 
                              'hazard':[np.mean(s.hazard) for s in np.split(hazard[:728], 728/7)]})
print(weekly_hazard.head(), weekly_hazard.tail())

sns.jointplot('weeks since origination', 'hazard', data=weekly_hazard, kind='reg',
              xlim=(-2, n_wks+2),  size=8)
sns.jointplot('weeks since origination', 'hazard', data=weekly_hazard, kind='resid',  
              ylim=(-.00025, .0006), size=8)
```
